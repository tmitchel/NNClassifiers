```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ltau2016 = pd.HDFStore('../datasets/stable_unscale.h5')['df']
ltau2016 = ltau2016[(ltau2016['sample_names'] != 'mc_ZTT')]
ltau2017 = pd.HDFStore('../datasets/ltau2017_unscale.h5')['df']
ltau2017 = ltau2017[(ltau2017['sample_names'] != 'mc_ZTT')]
```

```python
plt.figure(figsize=(20,16))

plt.subplot(2,2,1)
ltau2016['el_pt'].plot.hist(range=(50, 200), bins=50, label='2016', alpha=0.5, density=True)
ltau2017['el_pt'].plot.hist(range=(50, 200), bins=50, label='2017', alpha=0.5, density=True)
plt.yscale('log')
plt.title('electron pT')
plt.legend()

plt.subplot(2,2,2)
ltau2016['mu_pt'].plot.hist(range=(50, 200), bins=50, label='2016', alpha=0.5, density=True)
ltau2017['mu_pt'].plot.hist(range=(50, 200), bins=50, label='2017', alpha=0.5, density=True)
plt.yscale('log')
plt.title('muon pT')
plt.legend()

plt.subplot(2,2,3)
ltau2016['t1_pt'].plot.hist(range=(50, 200), bins=50, label='2016', alpha=0.5, density=True)
ltau2017['t1_pt'].plot.hist(range=(50, 200), bins=50, label='2017', alpha=0.5, density=True)
plt.yscale('log')
plt.title('tau pT')
plt.legend()

plt.subplot(2,2,4)
ltau2016['met'].plot.hist(range=(50, 200), bins=50, label='2016', alpha=0.5, density=True)
ltau2017['met'].plot.hist(range=(50, 200), bins=50, label='2017', alpha=0.5, density=True)
plt.yscale('log')
plt.title('MET')

plt.legend()
```
